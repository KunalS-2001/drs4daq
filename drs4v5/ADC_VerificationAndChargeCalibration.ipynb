#### **ADC integration**

```python
def ADC_integral(wf,tm,ch,trig,n_off,i_window,freq,f_edge):
    i=1
    t_sign=1
    if f_edge==False :
        t_sign=-1
    integ=0
    while(i<1024):
        if wf[ch][i]*t_sign < trig*t_sign:
            break
        i+=1
    start=i
    if(start>1023): 
        return 0,start,i
    while(i>0):
        if tm[ch][start]-tm[ch][i] > n_off:
            start=i
            break
        i-=1
    window=0
    while i<1024 and window<i_window:
        dt=tm[ch][i]-tm[ch][i-1]
        window+=dt
        integ+=dt*wf[ch][i]
        i+=1
    return integ,start,i       
```
